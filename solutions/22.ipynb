# 22. Generate Parentheses

Given `n` pairs of parentheses, write a function to generate all combinations
of well-formed parentheses.

## Example 1:

```bash
Input: n = 3
Output: ["((()))","(()())","(())()","()(())","()()()"]
```

## Example 2:

```bash
Input: n = 1
Output: ["()"]
```

## Constraints:

- `1 <= n <= 8`

````python
from typing import List, Tuple

from typing import List, Tuple

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:

        current = set( [("", 0)] )
        next = set()

        length = 0
        while length < 2 * n:
            for generated, balance in current:
                if balance > 0:
                    next.add( (f"{generated})", balance-1) )

                if balance < n:
                    next.add( (f"{generated}(", balance+1) )

            current = next
            next = set()
            length += 1

        final = [ generated for (generated, balance) in current if balance == 0 ]

        return final

    
    def generateParenthesis2(self, n: int) -> List[str]:

        def helper(generated: List[Tuple[str, int]], length: int):

            if length == 2 * n:
                final = [ parens for (parens, balance) in generated if balance == 0 ]
                return final


            next = []
            for par, balance in generated:
                if balance > 0:
                    next.append( (f"{par})", balance-1) )

                if balance < n:
                    next.append( (f"{par}(", balance+1) )

            return helper(next, length+1)
        
        return helper([("", 0)], 0)
    

````

````python
solver = Solution()

for i in range(1, 10):
    print(f"{i}: {solver.generateParenthesis(i)}")
````
